```python
import numpy as np
import pandas as pd
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv("https://github.com/anvarnarz/praktikum_datasets/raw/main/museum_visitors.csv", index_col=0)
df.head()
```

```python
df.shape
```

```python
sns.lineplot(data=df)
plt.show()
```

```python
plt.plot(df)
plt.show()
```

```python
plt.figure(figsize=(20,8))
plt.title("Muzeylarga tashrif grafigi")
plt.xlabel("Sana")
plt.ylabel("Odamlar soni")
plt.xticks(rotation=90)
sns.lineplot(data=df,linewidth=2,marker='o',markerfacecolor='b')
plt.grid()
plt.show()
```
