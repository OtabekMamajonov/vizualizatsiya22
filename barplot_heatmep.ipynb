```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/main/uzbekistan.csv", index_col=0)
df.columns = ['maydon','aholi']
df
```

```python
plt.figure(figsize=(18,9))
sns.barplot(x=df.index, y=df['aholi'])
plt.grid()
plt.title('aholi soni viloyatlar  boyicha taqsimoti')
plt.xlabel('viloyatlar')
plt.ylabel('aholi soni')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(18,9))
sns.barplot(x=df.index, y=df['maydon'])
plt.grid()
plt.title('maydon viloyatlar  boyicha taqsimoti')
plt.xlabel('viloyatlar')
plt.ylabel('maydoni')
plt.xticks(rotation=90)
plt.show()
```

```python
df1 = pd.read_csv("https://github.com/anvarnarz/praktikum_datasets/raw/main/museum_visitors.csv", index_col=0, parse_dates=[0], usecols=[0,1])
df1.head()
```

```python
df1['month'] = [i.month for i in df1.index]
df1['year'] = [i.year for i in df1.index]
df1.head()
```

```python
df1.groupby(['month','year']).max().head(10)
```

```python
df_m = df1.groupby(['month','year']).max().unstack(level=0)
df_m.head()
```

```python
plt.figure(figsize=(16,6))
sns.heatmap(data=df_m)
plt.show()
```

```python
plt.figure(figsize=(16,6))
sns.heatmap(data=df_m, cmap="YlGnBu", linewidths=1, square=True)
xticks_labels = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun',
                'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
plt.xticks(np.arange(12) + .5, labels=xticks_labels)
plt.show()
```
