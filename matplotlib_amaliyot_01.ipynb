```python
import pandas as pd
import numpy as np
```

```python
import ssl
ssl._create_default_https_context = ssl._create_unverified_context
```

```python
listlar = pd.read_html('https://en.tutiempo.net/climate/ws-384570.html',index_col=0)
df=listlar[3][listlar[3].index>1990]
df.head()
```

```python
nomlar = listlar[4]
nomlar.head()
```

```python
df.info()
```

```python
df = df.replace("-", np.nan)
```

```python
df = df.astype(dtype='float')
df.head(10)
```

```python
df.isnull().sum()/len(df)*100
```

```python
df.drop(['PP'],axis=1,inplace=True)
```

```python
df.fillna(method='ffill', inplace=True)
df
```

```python
df.to_csv("tashtemp.csv")
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.figure(figsize=(15,5))

plt.plot(df['T'],linestyle='--',color='g',marker='o',linewidth=3,markerfacecolor='r')
plt.title('ortacha yillik harorat grafigi')
plt.xlabel('yillar')
plt.ylabel('harorat')
plt.grid(linestyle='--',marker='o')
plt.show()
```

```python
plt.figure(figsize=(20,10))
plt.plot(df['Tm'],linewidth=2.5,linestyle='--',marker='o',markerfacecolor='g',color='b',label='ortacha miniman harorat')
plt.plot(df['TM'],linewidth=2.5,linestyle='--',marker='o',markerfacecolor='r',color='y',label='ortacha maksimal harorat')
plt.grid(linestyle='--')
plt.title('yillik max va mix harorat')
plt.xlabel('yil')
plt.ylabel('harorat')
plt.legend()
plt.savefig('tashminmax.png')

plt.show()
```
