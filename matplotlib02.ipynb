```python
x=np.arange(0, 4*np.pi,0.1)
sin=np.sin(x)
cos=np.cos(x)
```

```python
plt.figure(figsize=(20,10))

plt.plot(x,sin,
         linewidth=2,
         linestyle='--',
         color='r',
         alpha=0.7,
         marker='o')

plt.plot(x,cos,
         linewidth=2,
         linestyle='--',
         color='g',
         alpha=0.7,
         marker='o')

plt.title('sin va cos grafigi')
plt.xlabel('pi')
plt.legend(['sin(x)','cos(x)'],loc=0,fontsize='x-large')
plt.grid(color='tomato',alpha=0.5,linestyle='--')
plt.savefig('grafik1.png')
plt.show()
```
